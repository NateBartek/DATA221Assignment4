```python
from keras.layers import Dense
from tensorflow.keras.datasets import fashion_mnist
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers

tf.random.set_seed(42)
(x_train , y_train ) , (x_test , y_test ) = fashion_mnist.load_data ()


x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

x_train = x_train[..., None]
x_test = x_test[..., None]

print("Train shape:", x_train.shape)  # (60000, 28, 28, 1)
print("Test shape: ", x_test.shape)

cnn_model = Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2)),


    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2)),

    layers.Flatten(),

    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_model.summary()

cnn_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"] )

cnn_model.fit(x_train, y_train, validation_split=0.1, epochs=15, batch_size=64)

test_loss, test_acc = cnn_model.evaluate(
x_test, y_test, verbose=0)

print(f"Test Loss: {test_loss:.2f}")
print(f"Test accuracy:, {test_acc*100:.2f}")

# CNNs are generally preferred over fully connected networks for image data, because generally they would have to many parameters and would take up to much time. For example an RGB image of 64x64x3 would have 12,288 inputs and for the first hidden layer of 1000 hidden units there would be 12,288*1000 = 12,288,000 parameters and 1000 biases. This is an extreme case and would result in more parameters, more memory, slower training, and higher chance of overfitting.

# For the Fashion MNIST dataset the convolution layer takes a filter (kernel) over the pixels and is looking for distinct features such as the outline of a shoe, or the changing shades for a handbag.

```
