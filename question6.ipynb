```python
# imports
from keras.layers import Dense
from tensorflow.keras.datasets import fashion_mnist
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers

# creating a random seed so that I can reproduce the results
tf.random.set_seed(42)

# creating the training and testing split from the fashion mnist dataset
(x_train , y_train ) , (x_test , y_test ) = fashion_mnist.load_data ()

# standardizing/scaling my features
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# added channel dimension so that it is compatible with Conv2D Layers
x_train = x_train[..., None]
x_test = x_test[..., None]

# printing out the shapes of the x_train and x_test to confirm
print("Train shape:", x_train.shape)
print("Test shape: ", x_test.shape)


# creating the cnn Model based off of sequential
cnn_model = Sequential([

    # first convolution layer, with 32 filters of size 3x3
    # padding to ensure that it doesn't lose it size
    # the input shape is 28,28,1
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), padding='same'),
    # first maxPooling layer
    layers.MaxPooling2D(pool_size=(2, 2)),

    # second convolution layer with 64 filters of size 3x3
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    # second maxPooling layer
    layers.MaxPooling2D(pool_size=(2, 2)),

    # flattening the 2d features into the appropriate format (1D) for the dense layers
    layers.Flatten(),

    # two fully connected layers for learning complex non-linear relationships
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    # final output layer and it has 10 neurons for the 10 different classes
    Dense(10, activation='softmax')
])

# displays a summary of the cnn, which includes parameters and layers
cnn_model.summary()


# compiling the neural network
# sparse_categorical_crossentropy is used for multiclassification using integers
# the neural network metric is accuracy
# the optimizer is adam for efficiency
cnn_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"] )


# training the cnn model with the training data
# has 15 epochs so it goes over the training data 15 times
# the batch size specifies number of samples per gradient update
# the validation split is 10% meaning that 10% of the training data is used for validation
cnn_model.fit(x_train, y_train, validation_split=0.1, epochs=15, batch_size=64)

# evaluating the cnn model based off of the testing data
test_loss, test_acc = cnn_model.evaluate(
x_test, y_test, verbose=0)

# printing out the performance of the cnn model in a nice formatted way
print(f"Test Loss: {test_loss:.2f}")
print(f"Test accuracy:, {test_acc*100:.2f}")

# answers
# CNNs are generally preferred over fully connected networks for image data, because generally they would have to many parameters and would take up to much time. For example an RGB image of 64x64x3 would have 12,288 inputs and for the first hidden layer of 1000 hidden units there would be 12,288*1000 = 12,288,000 parameters and 1000 biases. This is an extreme case and would result in more parameters, more memory, slower training, and higher chance of overfitting.

# For the Fashion MNIST dataset the convolution layer takes a filter (kernel) over the pixels and is looking for distinct features such as the outline of a shoe, or the changing shades for a handbag.

```
