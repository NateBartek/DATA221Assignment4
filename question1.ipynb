```python
from sklearn.datasets import load_breast_cancer

# loading the dataset
breast_cancer_dataset = load_breast_cancer()

# setting my features to equal x and target to label y
X = breast_cancer_dataset.data
y = breast_cancer_dataset.target

# print the shapes of my features and labels
print(f"shape of the feature matrix, X: {X.shape}")
print(f"The shape of the target vector, y: {y.shape}")

# creating the variables that will store the amount of times that they occur
malignant = 0
benign = 0

# counting the amount of times that malignant and benign occur in the breast cancer dataset
for test in y:
    if test == 0:
        malignant += 1
    else:
        benign += 1

# printing out the number in a nice formatted way
print(f"The number of breast cancer in this dataset is: {malignant}")
print(f"The number of benign in this dataset is: {benign}")

# answers
# The dataset is imbalanced because there are more malignant cases than benign cases.
# While the imbalance isn't the worst it could be better as just this amount of imbalance can and will
# cause model performance issues.
#
# Class balance is important when considering for classification models because these algorithms want to maximize accuracy. Therefore, if the dataset isn't balanced then the model could then predict the majority
# class most of the time and still end up with achieving a high accuracy score because most of the time it will end up being the majority class. However, this isn't the goal as it will do a very poor job of correctly predicting the minority class.

```
